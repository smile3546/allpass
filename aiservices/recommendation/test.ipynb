```python
import pandas as pd
from common.utils.dbcon import engine
from sqlalchemy import text
```

```python
sql_query = """
    SELECT
    ts.trail_id,
    ts.trail_name_en,
    ts.filename,
    ts.distance,
    ts.elevation_range,
    ts.elevation_change,
    ts.elevation_gain,
    ts.elevation_loss,
    ts.high_elevation,
    ts.max_slope_percent,
    ts.max_slope_degrees,
    ts.slope_std_dev,
    ts.slope_variance,
    ts.slope_freq_dist
    FROM ml_features.trail_segments AS ts
    ORDER BY ts.trail_id
    """

df = pd.read_sql(sql=sql_query, con=engine)
```
